```python

import time
import pyautogui
import pygetwindow as gw



# Function to check if WhatsApp window is active
def is_whatsapp_window_active():
    if True:
        active_window = gw.getActiveWindow()
        if active_window is not None and "WhatsApp" in active_window.title:
            return True
        time.sleep(1)


def convert_excel_toimage():
    import pandas as pd
    import matplotlib.pyplot as plt
    df = pd.read_excel('output_data.xlsx', engine='openpyxl')


# Create a figure and axis for the table with a lower DPI (e.g., 100)
    fig, ax = plt.subplots(figsize=(6, 4), dpi=100)  # Reduced DPI
    ax.axis('off')  # Turn off the axis

# Create a table and add data from the DataFrame
    table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Increase the cell height of the header cells and add line breaks for header cells with a space
    for (i, j) , cell in table._cells.items():
     if i == 0 :  # Header row
        cell.set_height(0.2)  # Increase the height as needed
        cell.set_fontsize(60)
        cell.set_width(0.12)

        # Add a line break for header cells with a space
        cell_text = cell.get_text().get_text()
        if ' ' in cell_text:
            cell_text = cell_text.replace(' ', '\n')
            cell.get_text().set_text(cell_text)  # Apply line break
     else:
        cell.set_width(0.12)
        cell.set_fontsize(60)
        cell.set_text_props(weight='bold', color='black')  # Make values darker

     if j == 0 or j==1:
        cell.set_width(0.17)
     if i != 0:
        cell.set_fontsize(80)

     if i == len(df):  # Last row
        cell.set_facecolor('#87CEEB')  # Sky blue background for the last row
        cell.set_text_props(weight='bold', color='black')  # Black text for the last row

    for key, cell in table._cells.items():
       if key[0] == 0:  # Header row
        cell.set_text_props(weight='bold', color='w')  # Header text color and style
        cell.set_facecolor('#1f77b4')  # Header cell background color
        cell.set_edgecolor('k')  # Header cell border color
       else:
        cell.set_edgecolor('k')  # Non-header cell border color

    table.auto_set_font_size(False)
    table.scale(15, 17)  # Adjust the table size as needed

# Save the table as an image (e.g., PNG) with a specified DPI
    import random
    a=round(random.random(),2)
    plt.savefig(f'table_image{a}.png', bbox_inches='tight', pad_inches=0)
    print("finished image creation")
    return f'table_image{a}.png'




path=convert_excel_toimage()



# Function to perform actions in WhatsApp window
def perform_whatsapp_actions():

    pyautogui.click(x=294,y=1042)
    time.sleep(4)
    pyautogui.typewrite("whatsapp")
    time.sleep(2)
    pyautogui.press('enter')

    time.sleep(2)

    roll_over=0
    s=is_whatsapp_window_active()
    if is_whatsapp_window_active():
        pyautogui.click(x=297, y=185)
        roll_over+=1

    if (roll_over!=1):
        return "NOT"

    if is_whatsapp_window_active() and roll_over == 1:
        pyautogui.hotkey("ctrl", "a")
        time.sleep(1)
        roll_over+=1

    if (roll_over!=2):
        return "NOT"
    if is_whatsapp_window_active() and roll_over ==2:
        pyautogui.hotkey("backspace")
        roll_over+=1

    if (roll_over!=3):
        return "NOT"
    if is_whatsapp_window_active() and roll_over ==3:
        pyautogui.typewrite("8825985219")
        time.sleep(2)
        roll_over+=1

    if (roll_over!=4):
        return "NOT"
    print(roll_over)
    if is_whatsapp_window_active() and  roll_over ==4:
        if is_whatsapp_window_active():
            pyautogui.click(x=357, y=287)
            time.sleep(2)
            roll_over+=2

    if (roll_over!=6):
        return "NOT"

    if is_whatsapp_window_active() and roll_over==6:
        pyautogui.click(x=761, y=974)
        time.sleep(4)
        roll_over+=1

    if (roll_over!=7):
        return "NOT"
    if is_whatsapp_window_active() and  roll_over ==7:
        pyautogui.click(x=738, y=779)
        roll_over+=1

    if (roll_over!=8):
        return "NOT"

    if is_whatsapp_window_active() and roll_over==8:
        pyautogui.click(pyautogui.position())
        time.sleep(4)
        roll_over+=1

    if (roll_over!=9):
        return "NOT"
    if gw.getActiveWindow().title=="Open" and roll_over ==9:
        if  gw.getActiveWindow().title=="Open":
          pyautogui.click(x=1181,y=69)
          time.sleep(2)
          pyautogui.typewrite("C:/Users/USER")
          time.sleep(2)
          pyautogui.hotkey("enter")
          time.sleep(2)
          pyautogui.click(x=643,y=625)
          time.sleep(1)
          pyautogui.typewrite(path)
          roll_over+=1

    if gw.getActiveWindow().title=="Open":
        pyautogui.press("enter")
        time.sleep(5)

    if is_whatsapp_window_active():
        pyautogui.press("enter")
        time.sleep(1)
        return "Executed"
    else:
        return "Not executed"


s=perform_whatsapp_actions()
print(s)
```
